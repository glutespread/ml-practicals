```python
import numpy as np
import matplotlib.pyplot as plt

class GD:
    def __init__(self):
        self.coeff = [1,1]
        self.history = []
    
    def train(self,X,Y,lr,epochs):
        m = X.shape[0]
        print(m)
        for i in range (epochs):
            y_pred = self.coeff[0] + (self.coeff[1])*X
            error = Y-y_pred
            cost = (sum(error*error))/(2*m)
            self.history.append(cost)
            print('error{}'.format(cost))
            self.coeff[0] = self.coeff[0] - (lr/m)*(sum(error))
            self.coeff[1] = self.coeff[1] - (lr/m)*(sum(error*X))
        print(self.coeff[0],',', self.coeff[1])
    
    def test(self,x1):
        y1 = self.coeff[0]+self.coeff[1]*x1
        plt.scatter(X,Y)
        plt.plot(x1,y1,'g-')
        
```

```python
gd1 = GD()
X_train = np.array([42,37,30,50,43,47,46])
Y_train = np.array([173,145,123,201,175,188,186])
X_test = np.array([40,90,20,45])
gd1.train(X_train,Y_train,-0.001,1000)
gd1.test(X_test)
```
